```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("all_perth_310121.csv")
df.head()
```

```python
fig = plt.figure()
sns.ecdfplot(x=df['BUILD_YEAR'])
fig.show()
```

```python
yearAndDateSold = df["DATE_SOLD"].str.split('-', expand=True)
df["MONTH_SOLD"] = yearAndDateSold[0].astype("int64")
df["YEAR_SOLD"] = yearAndDateSold[1].astype("int64")

fig = plt.figure()
sns.scatterplot(data=df, x=df["YEAR_SOLD"], y=df["PRICE"])
fig.show()
```

```python
fig = plt.figure()
plt.bar(df["YEAR_SOLD"], df["PRICE"])
fig.show()
```

```python
fig = plt.figure()
plt.pie(df["BEDROOMS"].value_counts(), labels= df["BEDROOMS"].value_counts(), startangle=90, colors=plt.cm.Paired.colors)
fig.show()
```
